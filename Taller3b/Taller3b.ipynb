Ejercicio 1

$$x_1 + x_2 = 7$$
$$-2x_1 + 5x_2 = 0$$

* $X_0 = (0,0)$

### Gauss-Jacobi

```python
A_1 = [[1, 1], [-2, 5]]
b_1 = [7, 0]
n = len(b_1)

gauss_jacobi(A=A_1, b=b_1, x=[0] * n, tol=1e-5, max_iter=10) 
```

```python
import matplotlib.pyplot as plt

solucion, historial = gauss_jacobi(A=A_1, b=b_1, x=[0] * n, tol=1e-5, max_iter=10)

puntos = np.array(historial).squeeze()

def f1(x):
    return 7 - x
def f2(x):
    return (2/5)*x

def graficar_trayectoria_2d(puntos):

    x = np.linspace(-1, 8, 100)
    x1 = puntos[:, 0]
    x2 = puntos[:, 1]

    plt.figure(figsize=(8, 8))
    
    plt.plot(x, f1(x), label='$x_1 + x_2 = 7$', color='orange')
    plt.plot(x, f2(x), label='$-2x_1 + 5x_2 = 0$', color='purple')
    plt.plot(x1, x2, linestyle='--', marker='o', color='black', markerfacecolor='red', label='Trayectoria Gauss-Jacobi')

    plt.plot(x1[0], x2[0], marker='o', color='green', label='Punto inicial')

    plt.title('Trayectoria para el valor inicial $X_0 = (0,0)$')
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.legend()
    plt.axis('equal') 
    plt.show()

graficar_trayectoria_2d(puntos)
```

Ejercicio 3

$$A = \begin{bmatrix}1 & -1 & 2 & 0 \\\ -1 & 1 & -1 & 3 \\\ 2 & -1 & 1 & -1 \\\ 0 & 3 & -1 & 2 \end{bmatrix}$$

$$b = \begin{bmatrix}1360 \\\ 1130 \\\ 1350 \\\ 3650 \end{bmatrix}$$

* $X_0 = (1,2,3,4)$

### Gauss-Jacobi

```python
A_3 = [[1, -1, 2, 0], [-1, 1, -1, 3], [2, -1, 1, -1], [0, 3, -1, 2]]
b_3 = [1360, 1130, 1350, 3650]
n = len(b_3)

gauss_jacobi(A=A_3, b=b_3, x=[1, 2, 3, 4], tol=1e-5, max_iter=20)
```

```python
import matplotlib.pyplot as plt

solucion, historial = gauss_jacobi(A=A_3, b=b_3, x=[1, 2, 3, 4], tol=1e-5, max_iter=20)

puntos = np.array(historial).squeeze() 

def graficar_convergencia(puntos):
    num_iteraciones, num_variables = puntos.shape
    iteraciones = np.arange(num_iteraciones)

    plt.figure(figsize=(10, 6))

    for i in range(num_variables):
        plt.plot(iteraciones, puntos[:, i], 
                 linestyle='--',      
                 marker='o',          
                 label=f'Variable $x_{i+1}$')

    plt.title('Trayectoria de convergencia del Método de Gauss-Jacobi')
    plt.xlabel('Número de Iteración ($k$)')
    plt.ylabel('Valor de la Variable')
    plt.grid(True, linestyle=':', alpha=0.6)
    plt.legend()
    plt.show()

graficar_convergencia(puntos)
```

### Gauss-Seidel

```python
gauss_seidel(A=A_3, b=b_3, x=[1, 2, 3, 4], tol=1e-5, max_iter=20)
```
