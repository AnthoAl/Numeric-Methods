* Implemente el conteo de sumas y restas para ambos métodos

````python
import numpy as np
import matplotlib.pyplot as plt

n = np.linspace(-10, 10, 400)

y1 = (n**3 / 3) + (n**2 / 2) - (5*n / 6)
y2 = (n**3 / 2) - (n / 2)

plt.figure(figsize=(8, 6))

plt.plot(n, y1, label=r'Eliminación Gaussiana', color='blue', linewidth=2)

plt.plot(n, y2, label=r'Gauss-Jordan', color='red', linestyle='--', linewidth=2)

plt.title('Gráfica de las funciones de sumas y restas para ambos métodos', fontsize=14)
plt.xlabel('n', fontsize=12)
plt.ylabel('Sumas/Restas', fontsize=12)
plt.xlim(0, 5)
plt.ylim(-1, 20)
plt.axhline(0, color='black', linewidth=0.5) # Eje X
plt.axvline(0, color='black', linewidth=0.5) # Eje Y
plt.grid(True, linestyle=':', alpha=0.7)
plt.legend(fontsize=12)

plt.show()
````

* Compare la complejidad computacional del método de eliminación gaussiana vs Gauss-Jordan.

Con base en la gráfica obtenida, se puede observar que el método de eliminación gaussiana es mucho más eficiente computacionalmente que el método de Gauss-Jordan puesto que la función que representa el número de sumas y restas para la eliminación gaussiana crece más lentamente en comparación con la función correspondiente al método de Gauss-Jordan.
