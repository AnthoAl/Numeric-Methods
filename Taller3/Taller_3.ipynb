## Series de Taylor

Grafique las curvas de las series de Taylor de varios órdenes para los siguientes casos:

+ $\cos{(x)}, x_0 = 0$  

```python
import math
import numpy as np

def f(x):
  return np.cos(x)

def f1(x):
  return 1 + 0*x

def f2(x):
  return 1 - (1/2)*(x**2)

def f3(x):
  return 1 - (1/2)*(x**2)

def f4(x):
  return 1 - (1/2)*(x**2) + (1/24)*(x**4)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(-5, 5, 400)

y_f = [f(x) for x in x_values]
y_f1 = [f1(x) for x in x_values]
y_f2 = [f2(x) for x in x_values]
y_f3 = [f3(x) for x in x_values]
y_f4 = [f4(x) for x in x_values]

plt.figure(figsize=(10, 6))

plt.plot(x_values, y_f, label='f(x) = cos(x)', color='blue')
plt.plot(x_values, y_f1, label='$P_1$(x)', color='red', linestyle = 'dashed')
plt.plot(x_values, y_f2, label='$P_2$(x)', color='green', linestyle = 'dashed')
plt.plot(x_values, y_f3, label='$P_3$(x)', color='purple', linestyle = 'dashed')
plt.plot(x_values, y_f4, label='$P_4$(x)', color='orange', linestyle = 'dashed')

plt.title('Gráfica de las Funciones')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()

plt.xlim(-5,5)
plt.ylim(-2,2)

plt.show()
```



+ $\frac{1}{1-x}, x_0 = 0.5$  

```python
import math

def f(x):
  return 1/(1-x)

def f1(x):
  return 4*x

def f2(x):
  return 4*x + 8*(x-0.5)**2

def f3(x):
  return 4*x + 8*(x-0.5)**2 + 16*(x-0.5)**3

def f4(x):
  return 4*x + 8*(x-0.5)**2 + 16*(x-0.5)**3 + 32*(x-0.5)**4
```

```python
from matplotlib.lines import lineStyles
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(-5, 5, 200)

y_f = [f(x) for x in x_values]
y_f1 = [f1(x) for x in x_values]
y_f2 = [f2(x) for x in x_values]
y_f3 = [f3(x) for x in x_values]
y_f4 = [f4(x) for x in x_values]

plt.figure(figsize=(10, 6))

plt.plot(x_values, y_f, label='f(x) = $\\frac{1}{1-x}$', color='blue')
plt.plot(x_values, y_f1, label='$P_1$(x)', color='red', linestyle = 'dashed')
plt.plot(x_values, y_f2, label='$P_2$(x)', color='green', linestyle = 'dashed')
plt.plot(x_values, y_f3, label='$P_3$(x)', color='purple', linestyle = 'dashed')
plt.plot(x_values, y_f4, label='$P_4$(x)', color='orange', linestyle = 'dashed')

plt.title('Gráfica de las Funciones')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlim(-5,5)
plt.ylim(-2,5)
plt.show()
```



+ $\ln{(x)}, x_0 = 1$  

```python
import math

def f(x):
  return math.log(x)

def f1(x):
  return x - 1

def f2(x):
  return - ((x-1)**2)/2 + x - 1

def f3(x):
  return (1/3) * (x - 1)**3 - ((x-1)**2)/2 + x - 1

def f4(x):
  return - (1/4) * (x - 1)**4 + (1/3) * (x - 1)**3 - ((x-1)**2)/2 + x - 1
```

## POLINOMIOS DE LAGRANGE

Encuentre el polinomio de Lagrange para los siguientes datos y grafique:
+ $(0,0), (30,0.5), (60,\frac{\sqrt{3}}{2}), (90,1)$

**Solución**

Número de puntos 4 $\rightarrow{Orden = 3}$

Polinomios obtenidos para $L_k(x)$



$L_0 = \frac{(x^2-90x+1800)(x-90)}{162000}$

$L_1 = \frac{(x^2-60x)(x-90)}{54000}$

$L_2 = -\frac{(x^2-30x)(x-90)}{54000}$

$L_3 = \frac{(x^2-30x)(x-60)}{162000}$



Polinomio de Lagrange resultante:

$P_3 (x) = (0.5)*\frac{(x^2-60x)(x-90)}{54000} - \frac{\sqrt{3}}{2} * \frac{(x^2-30x)(x-90)}{54000} + \frac{(x^2-30x)(x-60)}{162000}$

```python
import math

def lagrange1(x):
  return (0.5)*(((x**2 - 60*x)*(x-90))/54000) - (math.sqrt(3)/2) * (((x**2-30*x)*(x-90))/54000) + (((x**2-30*x)*(x-60))/162000) 
```

```python
from matplotlib.lines import lineStyles
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(-5, 100, 400)

y_f = [lagrange1(x) for x in x_values]

x_puntos = [0, 30, 60, 90]
y_puntos = [0, 0.5, (math.sqrt(3)/2), 1]

puntos_a_etiquetar = {
    'A(0, 0)': (0, 0), 
    'B(30, 0.5)': (30, 0.5), 
    'C(60, 0.866)': (60, y_puntos[2]), 
    'D(90, 1)': (90, 1)                  
}

plt.figure(figsize=(10, 6))

plt.plot(x_values, y_f, label='$P_3$(x)', color='green', linestyle = 'dashed')

plt.plot(
    x_puntos, 
    y_puntos, 
    color='orange', 
    marker='o', 
    linestyle='',
    label='Puntos de Interpolación'
)

for label, (x, y) in puntos_a_etiquetar.items():
    plt.annotate(
        label,                      
        (x, y),                     
        textcoords="offset points",
        xytext=(0, 10),             
        ha='center',                
        fontsize=9,
        color='blue'
    )

plt.title('Gráfica Resultante')
plt.xlabel('x')
plt.ylabel('$P_3(x)$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.ylim(-1,5)
plt.legend()
plt.show()
```



+ $(−2,5),(1,7),(3,11),(7,34)$

**Solución**

Número de puntos 4 $\rightarrow{Orden = 3}$

Polinomios obtenidos para $L_k(x)$



$L_0 = -\frac{(x^2-4x+3)(x-7)}{135}$

$L_1 = \frac{(x^2-x-6)(x-7)}{36}$

$L_2 = -\frac{(x^2+x-2)(x-7)}{40}$

$L_3 = \frac{(x^2+x-2)(x-3)}{216}$



Polinomio de Lagrange resultante:

$P_3 (x) = -5 * \frac{(x^2-4x+3)(x-7)}{135} + 7 * \frac{(x^2-x-6)(x-7)}{36} - 11 * \frac{(x^2+x-2)(x-7)}{40} + \frac{(x^2+x-2)(x-3)}{216}$

```python
import math

def lagrange3(x):
  return (-5)*(((x**2-4*x+3)*(x-7))/135) + 7 * (((x**2-x-6)*(x-7))/36) - 11 * (((x**2+x-2)*(x-7))/40) + 34 * (((x**2+x-2)*(x-3))/216)
```

```python
from matplotlib.lines import lineStyles
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(-7, 10, 400)

y_f = [lagrange3(x) for x in x_values]

x_puntos = [-2, 1, 3, 7]
y_puntos = [5, 7, 11, 34]

puntos_a_etiquetar = {
    'A(-2, 5)': (-2, 5), 
    'B(1, 7)': (1, 7), 
    'C(3, 11)': (3, 11), 
    'D(7, 34)': (7, 34)  
}

plt.figure(figsize=(10, 6))

plt.plot(x_values, y_f, label='$P_3$(x)', color='green', linestyle = 'dashed')

plt.plot(
    x_puntos, 
    y_puntos, 
    color='orange', 
    marker='o', 
    linestyle='',
    label='Puntos de Interpolación'
)

for label, (x, y) in puntos_a_etiquetar.items():
    plt.annotate(
        label,                     
        (x, y),                    
        textcoords="offset points",
        xytext=(0, 10),            
        ha='center',               
        fontsize=9,
        color='blue'
    )

plt.title('Gráfica Resultante')
plt.xlabel('x')
plt.ylabel('$P_3(x)$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.ylim(-10,40)
plt.legend()
plt.show()
```

## Conclusiones

+ En series de Taylor, se identificó que todos los polinomios de grado n pasan por el valor de $x_0$ y cada vez que se alejan de este punto presentan mayor error con respecto a la función original, como se observa en la gráfica de cada uno de los ejercicios realizados.
+ Con respecto a los polinomios de Lagrange, se puede concluir que son útiles en cantidades pequeñas de puntos puesto que, si se tiene, por ejemplo, 500 puntos, se debería calcular un polinomio de orden 499 siendo este, un proceso demasiado extenso y poco eficiente. Además, una desventaja de esta técnica es que los puntos dados pueden presentar error, provocando que el polinomio resultante no represente de forma adecuada los valores reales.
